# APRIL 2017

```python
df = pd.read_excel('dxc_archive.xlsx')
df = pd.DataFrame(df)
df = df[['Project Manager','Expected Start','Duration','Project ID']]
df = df.sort_values(by='Expected Start', ascending=True)
df['Month'] = df['Expected Start'].astype(str)
df = df[['Project Manager','Month','Duration','Project ID']]

#Importing NNSR overview excel file

NSSROverview = pd.read_excel('NSSR Overview.xlsx')
NSSROverview_df = pd.DataFrame(NSSROverview)
NSSROverview_df = NSSROverview_df[['NSSR Project List', 'Request Type','Request Total Project Quoted Amount']]


#Creating a df that contains the projects that were carried over from 6 months before april 2017

df_1 = df.loc[df['Month'] == '2016-10-01']
df_2 = df.loc[df['Month'] == '2016-11-01']
df_3 = df.loc[df['Month'] == '2016-12-01']
df_4 = df.loc[df['Month'] == '2017-01-01']
df_5 = df.loc[df['Month'] == '2018-02-01']

df_1 = df_1.loc[df_1['Duration'] > 6]
df_2 = df_2.loc[df_2['Duration'] > 5]
df_3 = df_3.loc[df_3['Duration'] > 4]
df_4 = df_4.loc[df_4['Duration'] > 3]
df_5 = df_5.loc[df_5['Duration'] > 2]

previous = pd.concat([df_1, df_2, df_3, df_4, df_5])
```

```python
#Extracting a list that contains the list that contains the NSSR numbers obtained from the 
april_2017_raw = df.loc[df['Month'] == '2017-04-01']
april_2017_raw = pd.concat([april_2017_raw, previous])
april_2017 = april_2017_raw
april_2017 = april_2017.groupby("Project Manager")
PM_list = list(april_2017.groups)
april_2017 = april_2017.count()
april_2017.apply(list).apply(pd.Series)
april_2017 = april_2017[april_2017.Month == 2]
unwanted = ["Dhond; Utkarsh","Marlavadi Narayana; Diwakar","Palanivelu Kumersha; Karthikeyan"]
PM_list = [x for x in PM_list if x not in unwanted]
pm_rows = []
for pm in PM_list:
    pm_df = april_2017_raw.loc[april_2017_raw['Project Manager'] == pm]
    pm_rows.append(pm_df)
pm_df = pd.concat(pm_rows)
NSSR_numbers = list(pm_df['Project ID'])

i = 0
lengths = []
for index, row in NSSROverview_df.iterrows():
    lenght = len(NSSROverview_df['NSSR Project List'][i])
    lengths.append(lenght)
    i += 1
NSSROverview_df['NSSR_lenght'] = lengths
NSSROverview_df
NSSROverview_df = NSSROverview_df[NSSROverview_df.NSSR_lenght == 6]
NSSROverview_df["NSSR Project List"] = pd.Series(NSSROverview_df["NSSR Project List"], dtype='int32')

complexity_rows = []
for number in NSSR_numbers:
    NSSROverview_rows = NSSROverview_df.loc[NSSROverview_df['NSSR Project List'] == number]  
    complexity_rows.append(NSSROverview_rows)
complexity_df = pd.concat(complexity_rows)
complexity_df = complexity_df.drop('NSSR_lenght', axis = 1)
complexity_df.groupby('NSSR Project List').count().head()
```

```python
j = 0
for index, row in NSSROverview_df.iterrows():
    if (complexity_df['NSSR Project List'][j]) == 26612:
        complexity_df['Request Type'][j] = 'Detailed - High Complexity'
        #complexity_df.drop(complexity_df.index[j + 1])
        j += 1
    elif (complexity_df['NSSR Project List'][j]) == 26626:
        complexity_df['Request Type'][j] = 'Detailed - Medium Complexity'
        #complexity_df.drop(complexity_df.index[j])
        j += 1
    elif (complexity_df['NSSR Project List'][j]) == 26069:
        complexity_df['Request Type'][j] = 'Detailed - Medium Complexity'
        #complexity_df.drop(complexity_df.index[j + 1])
        j += 1
    else:
        break
        
complexity_df.groupby('Request Type').count()
```

```python
df = pd.read_excel('dxc_archive.xlsx')
df = pd.DataFrame(df)
df = df[['Project Manager','Expected Start','Duration','Project ID']]
df = df.sort_values(by='Expected Start', ascending=True)
df['Month'] = df['Expected Start'].astype(str)
df = df[['Project Manager','Month','Duration','Project ID']]

#Importing NNSR overview excel file

NSSROverview = pd.read_excel('NSSR Overview.xlsx')
NSSROverview_df = pd.DataFrame(NSSROverview)
NSSROverview_df = NSSROverview_df[['NSSR Project List', 'Request Type','Request Total Project Quoted Amount']]


#Creating a df that contains the projects that were carried over from 6 months before may 2017

df_1 = df.loc[df['Month'] == '2016-10-01']
df_2 = df.loc[df['Month'] == '2016-11-01']
df_3 = df.loc[df['Month'] == '2016-12-01']
df_4 = df.loc[df['Month'] == '2017-01-01']
df_5 = df.loc[df['Month'] == '2017-02-01']

df_1 = df_1.loc[df_1['Duration'] > 11]
df_2 = df_2.loc[df_2['Duration'] > 10]
df_3 = df_3.loc[df_3['Duration'] > 9]
df_4 = df_4.loc[df_4['Duration'] > 8]
df_5 = df_5.loc[df_5['Duration'] > 7]

previous = pd.concat([df_1, df_2, df_3, df_4, df_5])

#Extracting a list that contains the list that contains the NSSR numbers obtained from the 
april_2017_raw = df.loc[df['Month'] == '2017-04-01']
may_2017_raw = df.loc[df['Month'] == '2017-05-01']
june_2017_raw = df.loc[df['Month'] == '2017-06-01']
july_2017_raw = df.loc[df['Month'] == '2017-07-01']
august_2017_raw = df.loc[df['Month'] == '2017-08-01']
september_2017_raw = df.loc[df['Month'] == '2017-09-01']
september_2017_raw = pd.concat([april_2017_raw.loc[april_2017_raw['Duration'] > 5],
                        may_2017_raw.loc[may_2017_raw['Duration'] > 4], 
                        june_2017_raw.loc[june_2017_raw['Duration'] > 3],
                        july_2017_raw.loc[july_2017_raw['Duration'] > 2],
                        august_2017_raw.loc[august_2017_raw['Duration'] > 1],
                        september_2017_raw,
                        previous
                      ])
september_2017 = september_2017_raw
september_2017 = september_2017.groupby("Project Manager")
PM_list = list(september_2017.groups)
september_2017 = september_2017.count()
september_2017.apply(list).apply(pd.Series)
september_2017 = september_2017[september_2017.Month == 2]
unwanted = ['Hacker; Angel','Jow; Lily','Lew; William','Orozco; Raul','Palanivelu Kumersha; Karthikeyan','Sahtan; Adnan',
           'Ortega Cisneros; Tanya Cecilia']
PM_list = [x for x in PM_list if x not in unwanted]
pm_rows = []
for pm in PM_list:
    pm_df = september_2017_raw.loc[september_2017_raw['Project Manager'] == pm]
    pm_rows.append(pm_df)
pm_df = pd.concat(pm_rows)
NSSR_numbers = list(pm_df['Project ID'])

i = 0
lengths = []
for index, row in NSSROverview_df.iterrows():
    lenght = len(NSSROverview_df['NSSR Project List'][i])
    lengths.append(lenght)
    i += 1
NSSROverview_df['NSSR_lenght'] = lengths
NSSROverview_df
NSSROverview_df = NSSROverview_df[NSSROverview_df.NSSR_lenght == 6]
NSSROverview_df["NSSR Project List"] = pd.Series(NSSROverview_df["NSSR Project List"], dtype='int32')

complexity_rows = []
for number in NSSR_numbers:
    NSSROverview_rows = NSSROverview_df.loc[NSSROverview_df['NSSR Project List'] == number]  
    complexity_rows.append(NSSROverview_rows)
complexity_df = pd.concat(complexity_rows)
complexity_df = complexity_df.drop('NSSR_lenght', axis = 1)
complexity_df.groupby('NSSR Project List').count()

j = 0
for index, row in NSSROverview_df.iterrows():
    if (complexity_df['NSSR Project List'][j]) == 26612:
        complexity_df['Request Type'][j] = 'Detailed - High Complexity'
        #complexity_df.drop(complexity_df.index[j + 1])
        j += 1
    elif (complexity_df['NSSR Project List'][j]) == 272476:
        complexity_df['Request Type'][j] = 'Detailed - High Complexity'
        #complexity_df.drop(complexity_df.index[j + 1])
        j += 1
    elif (complexity_df['NSSR Project List'][j]) == 26626:
        complexity_df['Request Type'][j] = 'Detailed - Medium Complexity'
        #complexity_df.drop(complexity_df.index[j])
        j += 1
    elif (complexity_df['NSSR Project List'][j]) == 26069:
        complexity_df['Request Type'][j] = 'Detailed - Medium Complexity'
        #complexity_df.drop(complexity_df.index[j + 1])
        j += 1
    else:
        break
        
complexity_df.groupby('Request Type').count()
```
